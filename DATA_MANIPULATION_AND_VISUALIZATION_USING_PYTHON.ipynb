```python
# exploratory data analysis
# annot-disply numbers
# cmap-Blues,plasma,viridis
sns.heatmap(titanic.corr(numeric_only=True),annot=True,cmap='plasma')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# corrilation value in between -1 to 1
sns.scatterplot(x=titanic['age'],y=titanic['survived'])
```

```python
sns.scatterplot(x='age',y='survived',data=titanic,alpha=0.6,hue='sex')
```

```python
sns.stripplot(x='age',y='survived',data=titanic,alpha=0.6,hue='sex')
```

```python
sns.swarmplot(x='survived',y='age',data=titanic,hue='sex',dodge=True)
```

```python
sns.histplot(x='age',data=titanic,alpha=0.6,hue='sex',kde=True)
```

```python
sns.barplot(x='survived',y='age',data=titanic,alpha=0.6,hue='sex')
```

```python
sns.boxplot(x='survived',y='age',data=titanic,hue='sex')
```

```python
plt.hist(titanic['age'],bins=40)
plt.title=('Histogram')
plt.xlabel('age')
plt.legend()
plt.show()
```

```python
plt.hist(titanic['age'].dropna(),bins=40,color='skyblue',edgecolor='black',label='age')
plt.title=('Histogram of passanger ages')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
titanic['survived'].value_counts().plot.pie(autopct='%1.1f%%',colors=['lightcoral','lightgreen'])
plt.ylabel('')
plt.show()
```
